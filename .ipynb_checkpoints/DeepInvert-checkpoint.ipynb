```python
import torch.nn as nn
import torch.optim as optim
from torchvision import models
import torch.nn.functional as F
from collections import OrderedDict
from read_image import read_image, tensor_to_image
from torchvision.utils import make_grid, save_image
import matplotlib.pyplot as plt
```

```python
cnn = models.alexnet(pretrained=True)
cnn.eval()
```

```python
for name, param in cnn.named_parameters():
    if param.requires_grad:
        param.requires_grad = False
        print(name)
```

```python
image = read_image('./test/cat.jpg')
```

```python
target = cnn(image)
```

```python
target = target.detach()
```

```python
x = torch.rand_like(image,requires_grad= True)
```

```python
loss = nn.MSELoss()
```

```python
optimizer = optim.Adam([x], lr=0.1)
```

```python
for i in range(100):
    optimizer.zero_grad()
    out = cnn(x)
    mse = loss(out, target) / torch.norm(target) + 0.01*torch.sum(x**6)
    print(i, mse)
    mse.backward()
    optimizer.step()
```
